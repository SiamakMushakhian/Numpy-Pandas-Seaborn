## 2. Creating Figures and Subplots

**Subplots** are used to **display multiple plots** in **one figure**. For example, to **compare** them with each other.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

In **matplotlib**, **all plots** are located in a **figure object** whether it is a **one plot** or **mutiple subplot**.

**figure()**: To **create** a **figure object**

```python
fig = plt.figure()
```

We can **set** the **size** of the **figure object** using the argument **figsize = (8,6)**:

```python
fig = plt.figure(figsize=(8, 6))
```

There are two ways to have multiple subplots:
- plt.figure().add_subplot(nrows, ncolumns, index): **Create** a **figure object** and then **add subplot**.
- plt.subplots(nrows, ncolumns): **Directly creates figure** and **subplots**.

**add_subplot()**: To **add** a **subplot**. 
- We can **add multiple subplot**.
- We should **create** the **figure object** again because **add_subplot()** should be in the **same cell** as the **figure object**.

```python
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax4 = fig.add_subplot(2, 2, 4)
```

we can **create** a **plot** in one of these **subplots** using the **name** of the **subplot** and the **plot()** function:

```python
fig = plt.figure(figsize = (6, 4))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, (3, 4)) 
ax1.plot([1, 2, 3, 4, 5])
ax2.plot(np.random.randn(20).cumsum())
```

```python
fig = plt.figure()
fig.add_subplot(231)
ax1 = fig.add_subplot(2, 3, 1)            # equivalent but more general
fig.add_subplot(232, frameon=False)       # subplot with no frame
fig.add_subplot(233, projection='polar')  # polar subplot
fig.add_subplot(234, sharex=ax1)          # subplot sharing x-axis with ax1
fig.add_subplot(235, facecolor="red")     # red subplot
ax1.remove()                              # delete ax1 from the figure
```

Another method to **create subplots** is to use the function **plt.subplots**

```python
fig, ax = plt.subplots(3, 3)
```

```python
fig, ax = plt.subplots(3, 3, figsize=(8, 6))
```

To **create** a **plot** inside one of these subplots, we use the **name** of the **subplot** and the **plot()** function:

```python
fig, ax = plt.subplots(3, 3, figsize =(8, 6))
ax[0,0].plot([1, 2, 3, 4, 5])
ax[2,2].plot([1, 2, 3, 4, 5])
```

**subplots_adjust()**: To **adjust** the **spaces** around the **subplot** by using these arguments:
- **hspace** : to **adjust** the **height** between the subplots
- **wspace** : to **adjust** the **width** between the subplots

```python
fig, ax = plt.subplots(3, 3, figsize=(8, 6))
ax[0,0].plot([1, 2, 3, 4, 5])
ax[2,2].plot([1, 2, 3, 4, 5])
fig.subplots_adjust(hspace=0.4, wspace=0.4)
```

## 4. Customizing Ticks and Labels

Matplotlib has many methods to set the **range**, the **location**, and the **labels** of the **ticks**.

```python
data = np.random.randn(200).cumsum()
```

```python
fig, ax = plt.subplots()
ax.plot(data)
```

We can **change** the **ticks** of the **x-axis** to a any range of numbers using the function **set_xticks()**:

```python
fig, ax = plt.subplots()
ax.plot(data)
ticks = ax.set_xticks(range(0, 201, 50))
```

We can also pass the **ticks** in a simple **list**

```python
fig, ax = plt.subplots()
ax.plot(data)
ticks = ax.set_xticks([0, 50, 100, 150, 200])
```

We can also use **labels** of the **ticks** instead of numbers using the function **set_xticklabels()**. Note that the **ticks** should be defined **before labels** and the **number** of **elements** should be the **same**.

```python
fig, ax = plt.subplots()
ax.plot(data)
ticks = ax.set_xticks([0, 50, 100, 150, 200])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'])
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(data)
ticks = ax.set_xticks([0,50,100,150,200])
labels = ax.set_xticklabels(['one year sales','two years sales','three years sales',
                             'four years sales','five years sales'])
```

We can **change** the **orientation** of the **ticks** using the argument **rotation =** 60, this sets the orientation to 60 degrees:

```python
fig, ax = plt.subplots(1,1)
ax.plot(data)
ticks = ax.set_xticks([0,50,100,150,200])
labels = ax.set_xticklabels(['one year sales','two years sales','three years sales',
                             'four years sales','five years sales'], rotation=45)
```

To add a **label** for the **x-axis**, we use the function **set_xlabel()**:

```python
fig, ax = plt.subplots(1,1)
ax.plot(data)
ticks = ax.set_xticks([0,50,100,150, 200])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'])
ax.set_xlabel('Cumulative Sum')
```

We can **add** a **title** for the figure using the function **set_title()**

```python
fig, ax = plt.subplots(1,1)
ax.plot(data)
ticks = ax.set_xticks([0,50,100,150, 200])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'])
ax.set_xlabel('Cumulative Sum')
ax.set_title('Cumulative Sum for Random Numbers')
```

All methods for the x-axis works as well for the y-axis, just replace x with y.

```python
fig, ax = plt.subplots(1,1)
ax.plot(data)
ticks = ax.set_xticks([0,50,100,150, 200])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'])
ax.set_xlabel('Cumulative Sum')
ax.set_title('Cumulative Sum for Random Numbers')
ax.set_ylabel('Values')
```

## 5. Adding Legends

We use **legends** when we plot **multiple data** on the **same plot**, or **multiple subplots** on the **same figure**.

**Legends** are used for **identifying plot components** (which graph belongs to which dataset).

To **add legends** in **matplotlib**, we just add a **label** argument in the **plot()** function.

```python
data1 = np.random.randn(200).cumsum()
data2 = np.random.randn(200).cumsum()
data3 = np.random.randn(200).cumsum()
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(data1)
ax.plot(data2)
ax.plot(data3)
```

To **add** a **legend**, first we need to add a **label** for each **plot** using the argument **label = ' '**. The next step for adding a legend is to use the function **legend()**

```python
fig, ax = plt.subplots(1, 1)
ax.plot(data1, label='data1')
ax.plot(data2, label='data2')
ax.plot(data3, label='data3')
ax.legend()
```

We can **change** the **location** of the **legend** by using the argument **loc = ''**:

        ===============   =============
        Location String   Location Code
        ===============   =============
        'best'            0
        'upper right'     1
        'upper left'      2
        'lower left'      3
        'lower right'     4
        'right'           5
        'center left'     6
        'center right'    7
        'lower center'    8
        'upper center'    9
        'center'          10
        ===============   =============
        
**loc = 'best'**, put the legend **automatically** in an **empty corner**.

```python
fig, ax = plt.subplots(1, 1)
ax.plot(data1, label='data1')
ax.plot(data2, label='data2')
ax.plot(data3, label='data3')
ax.legend(loc='upper center')
```

## 6. Adding Texts and Arrows on a Plot

**Texts** and **arrows** enhance the quality of the figures for **presentations** and for **publications**.

**Texts** are added using the function **text()**

**Arrows** are added using the function **arrow()**

```python
x = np.arange(0, 10, 0.1)
y = np.sin(x)
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, y)
ax.text(3.5, 0.7, 'y = sin(x)', fontsize=15)
```

The **text** is **positioned** based on **x** and **y coordinates**.

The **text** can be also **located** based on **figure rectangle coordinates**, here **(0.5, 0.5)** means in the **center** of the **plot**:

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, y)
ax.text(0.38, 0.5, 'y = sin(x)', fontsize=15, transform=ax.transAxes)
```

We can **add arrows** to the plot, using the function **arrow()**:

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, y)
ax.arrow(0.5, 1, 0.6, 0, width=0.05)
ax.arrow(2.1, 1, 2.5, -1.85, width=0.05)
```

## 7. Adding Annotations and Drawings

**Annotations** is used in figures to draw attention to **special features** or **important points** on the graph.

```python
x = np.arange(6)
y = np.array([23, 34, 65, 78, 51, 55])
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, y, 'ko--')
ax.annotate('Maximum', (x[3], y[3])) # The text is placed at *xy*.
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, y, 'ko--')
ax.annotate('Maximum', xy=(x[3], y[3]), xytext=(3, 50)) # The text can be displayed in another position *xytext*.
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, y, 'ko--')
ax.annotate('Maximum', xy=(x[3], y[3]), xytext=(3, 50), arrowprops=dict(facecolor='green', shrink=0.1))
ax.annotate('Minimum', xy=(0, 23), xytext=(0, 50), arrowprops=dict())
# An arrow pointing from the text to the annotated point *xy* can then be added by defining *arrowprops*.
```

**Note**: The x and y should be in the range of data.

**Adding** a specific **shape** to a **plot** is a very simple task, but it consists of two steps:
- **Create** the **shape object**
- **Attach** the **object** on the **plot**

```python
fig, ax = plt.subplots(1, 1)
c = plt.Circle((0.6, 0.4), 0.2, fc='y')
ax.add_patch(c)
r = plt.Rectangle((0.3, 0.7), 0.5, 0.2, color='y', alpha= 0.5) # alpha = transparancy
ax.add_patch(r)
tri = plt.Polygon([[0.25, 0.25], [0.55, 0.5], [0.4, 0.7]], color='g', alpha=0.5)
ax.add_patch(tri)
```

## 12. Histograms and Density Plots

- A **histogram** is a type of **bar plots** that displays **frequency of values**.
- The data is **split** into **discrete ranges** (**intervals**) and each bar represents how many data points fall in that range.
- A **histogram** is used as a representation of the **distribution** of the **data** as it dispalys the **shape** and the **spread** of continuous data.
- A **histogram** can be created in **pandas** using as method called **plot.hist()**.
- **Before creating** a **histogram**, we need first to **decide** the **number of bins**.
- **Bins** in histograms are the **number of equally-spaced intervals** that are used to sort the data.
- **Changing** the **number of bins** will **change** the **shape** of the **histogram**.
- Usually, the **number of bins** lies **between 5 and 20**, and it **depends** on the **size** and the **range** of the data.

```python
df = pd.read_csv('data/ex13.csv')
df.head()
```

```python
df['age'].plot.hist(bins=12)
```

- We use the argument **color** to change the **color** of the **bars**.
- We use the argument **edgecolor** to change **border color** of the **bars**.
- We use the argument **linewidth** to change the **width** of the **borders lines**.

```python
df['age'].plot.hist(bins=12, color='orange', edgecolor='red', linewidth=1)
```

- **Density plot** is used to **visualize** the **distribution** of the **data**.
- The **advantage** of **density plot** over **histogram** is that the density plots are **not affected** by the **number of bins**.
- To **create** a **density plot**, we use the function **plot.density()**

```python
df['wage'].plot.density()
```

**Seaborn** can also be used to create **histograms** by using the function **sns.displot()**:

```python
sns.displot(df['wage'], bins=12, color='r')
```

**Seaborn** can also be used to create **density plots** by using the function **sns.displot()** and the argument (**kind = 'kde'**):

```python
sns.displot(df['wage'], kind='kde', color='green')
```

## 13. Scatter Plots and Pair Plots

- **Scatter** and **pair plots** are very common in **data science** and **statistical modeling**.
- **Scatter plots** are very useful to explore the **relationship between two data series**.

```python
df = pd.read_csv('data/boston.csv')
df.head()
```

We can use **seaborn** to create a **scatter plot** between two columns using the function **sns.regplot()**:

```python
sns.regplot(x=df['medv'], y=df['crim'])
```

To **remove** the **confidence interval** ( the **light blue channel** around the **regression line**) we use the argument **ci = False**

```python
sns.regplot(x=df['medv'], y=df['crim'], ci=False)
```

To **remove** thre **regression line** from the **scatter plot**, we use the argument **fit_reg = False**

Using **seaborn**, we can create a **matrix** of **scatter plots** among **all columns** using the function **sns.pairplot()**:

```python
sns.pairplot(df)
```

**Note**: The **diagonal plots** are the **histogram** for each variable which enable us to check the **distribution** of that variable.

It is **usuall** among analysts to replace the histograms in the diagonal plots in the scatter plot matrix with **density plots**, and we do it using the argument **diag_kind = 'kde'**:

```python
sns.pairplot(df, diag_kind='kde')
```

## 14. Factor Plots for Categorical Data

- **Line** and **scatter plots** are used to **visualize continuous data**.
- **Factor plots** are used to **visualize categorical data**.
- In **seaborn**, **factor plots** are created using the function **catplot()**.

```python
df = pd.read_csv('data/credit.csv')
df.head()
```

```python
sns.catplot(x='Gender', y='Income', data=df)
```

In **factor plots**, we can use **bars** instead of dots by using the argument **kind = 'bar'**

```python
sns.catplot(x='Gender', y='Income', data=df, kind='bar')
```

```python
sns.catplot(x='Gender', y='Income', data=df, kind='bar', ci=False)
```

We can also use **box plots** for a **categorical variable** to provide additional statistics using the argument (**kind = 'box'**)

```python
sns.catplot(x='Gender', y='Income', data=df, kind='box')
```

We can also add a **hue** variable using the argument hue, in this example to explore the **relationship** between ethnicity and income with respect to the hue variable 'Gender'.

```python
sns.catplot(x='Ethnicity', y='Income', hue='Gender', data=df, kind='bar')
```
