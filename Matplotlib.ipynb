## 2. Creating Figures and Subplots

**Subplots** are used to **display multiple plots** in **one figure**. For example, to **compare** them with each other.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

In **matplotlib**, **all plots** are located in a **figure object** whether it is a **one plot** or **mutiple subplot**.

**figure()**: To **create** a **figure object**

```python
fig = plt.figure()
```

We can **set** the **size** of the **figure object** using the argument **figsize = (8,6)**:

```python
fig = plt.figure(figsize=(8, 6))
```

There are two ways to have multiple subplots:
- plt.figure().add_subplot(nrows, ncolumns, index): **Create** a **figure object** and then **add subplot**.
- plt.subplots(nrows, ncolumns): **Directly creates figure** and **subplots**.

**add_subplot()**: To **add** a **subplot**. Note that we can **add multiple subplot**.

**Note**: We should **create** the **figure object** again because **add_subplot()** should be in the **same cell** as the **figure object**.

```python
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)
```

we can **create** a **plot** in one of these **subplots** using the **name** of the **subplot** and the **plot()** function:

```python
fig = plt.figure(figsize = (6, 4))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, (3, 4)) 
ax1.plot([1, 2, 3, 4, 5])
ax2.plot(np.random.randn(20).cumsum())
```

```python
fig = plt.figure()
fig.add_subplot(231)
ax1 = fig.add_subplot(2, 3, 1)            # equivalent but more general
fig.add_subplot(232, frameon=False)       # subplot with no frame
fig.add_subplot(233, projection='polar')  # polar subplot
fig.add_subplot(234, sharex=ax1)          # subplot sharing x-axis with ax1
fig.add_subplot(235, facecolor="red")     # red subplot
ax1.remove()                              # delete ax1 from the figure
```

Another method to **create subplots** is to use the function **plt.subplots**

```python
fig, ax = plt.subplots(3, 3)
```

```python
fig, ax = plt.subplots(3, 3, figsize=(8, 6))
```

To **create** a **plot** inside one of these subplots, we use the **name** of the **subplot** and the **plot()** function:

```python
fig, ax = plt.subplots(3, 3, figsize =(8, 6))
ax[0,0].plot([1, 2, 3, 4, 5])
ax[2,2].plot([1, 2, 3, 4, 5])
```

**subplots_adjust()**: To adjust the spaces around the subplot by using these arguments:
- **hspace** : to **adjust** the **height** between the subplots
- **wspace** : to **adjust** the **width** between the subplots

```python
fig, ax = plt.subplots(3, 3, figsize=(8, 6))
ax[0,0].plot([1, 2, 3, 4, 5])
ax[2,2].plot([1, 2, 3, 4, 5])
fig.subplots_adjust(hspace=0.4, wspace=0.4)
```

```python
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
```

```python
# Create just a figure and only one subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')
```

```python
# Create two subplots and unpack the output array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)
```

```python
# Create four polar axes and access them through the returned array
fig, axs = plt.subplots(2, 2, subplot_kw=dict(projection="polar"))
axs[0, 0].plot(x, y)
axs[1, 1].scatter(x, y)
```

```python
# Share a X axis with each column of subplots
plt.subplots(2, 2, sharex='col')
```

```python
# Share a Y axis with each row of subplots
plt.subplots(2, 2, sharey='row')
```

```python
# Share both X and Y axes with all subplots
plt.subplots(2, 2, sharex='all', sharey='all')

# Note that this is the same as
#plt.subplots(2, 2, sharex=True, sharey=True)
```
